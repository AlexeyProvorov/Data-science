# **Seaborn и небольшое исследование**



```
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(style='ticks')
sns.set(rc={'figure.figsize':(8,5)})
data = sns.load_dataset('mpg')
data.dropna(inplace=True)
```

```
data.info()
```

```
sns.relplot(x='horsepower', y='mpg', data=data)
# The more horsepower, the less fuel consumption(mpg)
```

```
sns.countplot(x='origin', data=data)
# (origin)the largest number of car brands in the United States
```

```
sns.barplot(x='model_year', y='mpg', hue='origin', data=data) 
# tendency of increasing fuel consumption(mpg) in machines in other country
```

```
sns.boxplot(x='mpg', y='model_year', orient = 'h', data=data)
```

```
sns.violinplot(data=data, x='mpg', y='model_year', orient='h' )
```

```
sns.displot(data['mpg'], rug=True)
```

```
sns.lmplot(x='weight', y='mpg', data=data)
sns.lmplot(x='weight',y='mpg',hue='origin' ,data=data)
sns.lmplot(x='weight',y='mpg',col='origin' ,data=data)
```

```
sns.pairplot(data[data.origin=='usa'].drop(columns=['origin', 'name']))
```

```
sns.heatmap(data.groupby(["origin", 'cylinders'])['mpg'].mean().unstack().fillna(0))
```

```
sns.heatmap(data.groupby(['model_year', 'cylinders'])['mpg'].mean().unstack().fillna(0))
```
