```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import mglearn 
```

```python
from IPython.display import display
plt.rc('font', family='Verdana')

from sklearn.datasets import load_iris
iris_dataset=load_iris()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python

print("Ключи iris_dataset: \n{}".format(iris_dataset.keys()))
print ("info: {}".format(iris_dataset['filename']))

print("share array X_train:{}".format(X_train.shape))
print("share array y_train:{}".format(y_train.shape))

print("share array X_test:{}".format(X_test.shape))
print("share array y_test:{}".format(y_test.shape))
```

```python
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
grr = pd.plotting.scatter_matrix(iris_dataframe,c=y_train, figsize=(15,15), marker ='q',
hist_kwds = {'bins':20}, s=50, alpha=.70, cmap=mglearn.cm3)
```

метод k ближайших соседей


```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
print("формамассиваX_new: {}".format(X_new.shape))
```

```python
prediction = knn.predict(X_new)
print("Прогноз: {}".format(prediction))
print("Спрогнозированнаяметка: {}".format(iris_dataset['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print("Прогнозы для тестового набора:\n{}".format(y_pred))
```

```python
print("Правильность на тестовом наборе: {:.2f}".format(np.mean(y_pred== y_test)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print("Правильностьнатестовомнаборе: {:.2f}".format(knn.score(X_test, y_test)))
```
